```python
data=[np.random.normal(0,std,100) for std in range(1,4)]
plt.boxplot(data)
```

```python
labe="python","c++","javascript","c","java"
sizes=[70,10,35,29,50]
colors=["red","green","white","blue","yellow"]
explode=(0.1,0,0,0,0)

plt.pie(sizes,explode=explode,labels=labe,colors=colors,
autopct='%1.1f%%',shadow=True)
plt.axis('equal')
plt.show()
```

```python
sns.heatmap(df.corr()) # heatmap is used to analys how the features are correlated to each other.
```

# univariate graph analysis

```python
sns.jointplot(data=df)
```

# Pairplot : ## its used  when the independent features are more than two , then we will use this plot to see the relation.

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='sex')
```

#Displot : its used to identify the distribution of the features  

```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'],kde=False)
```

```python
sns.distplot(df['tip'],kde=True)
```

```python
sns.distplot(df['tip'],kde=True,bins=10)
```

# categorical plots 
-- its used to draw the graph of categorical data 

```python
# countplot:
--Show the counts of observations in each categorical bin using bars.
```

```python
sns.countplot('sex',data=df)
```

```python
sns.countplot('smoker',data=df)
```

```python
sns.countplot(y='sex',data=df)
```

```python
sns.countplot('tip',data=df) # not useful in numerical data.
```

```python
#barplot (combination of numerical and categorical)
###Show point estimates and confidence intervals as rectangular bars
```

```python
sns.barplot(x='sex',y='tip',data=df)
```

# Boxplot

```python
##Draw a box plot to show distributions with respect to categories.
```
